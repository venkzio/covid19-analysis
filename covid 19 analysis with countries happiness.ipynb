```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sea
```

# section 1

```python
df = pd.read_csv("time_series_covid_19_confirmed.csv")
```

```python
del df["Lat"]
```

```python
del df["Long"]
```

```python
df.head(2)
```

```python
df.shape
```

```python
groupe_df = df.groupby("Country/Region").agg(sum)

#groupe_df = groupe_df.agg(sum)
```

```python
pd.set_option("display.max_rows",None)
groupe_df
```

# section 2

```python
kk=groupe_df.iloc[:,2:].diff(axis = 1).max(axis = 1)
```

```python
kk.tail()
```

```python
groupe_df["Max_Infection_Rate"]=kk
```

```python
pd.set_option("display.max_rows",None)

groupe_df
```

```python
corona_data = groupe_df[["Max_Infection_Rate"]]
corona_data
```

```python
corona_data.head()
```
